```python
## Creating classes and objects

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person("Khem", 24)

person
```

```python
## Access modifiers (used for encapsulation)

class Person:
    def __init__(self, name, age, gender):
        self.__name = name # Private variable (name mangled to discourage external access)
        self._age = age # Protected variable (convention: intended for class and subclasses)
        self.gender = gender # Public variable (accessible everywhere)

    def __repr__(self):
        return f"Person(name={self.__name}, age={self._age}, gender={self.gender})"

person = Person("Khem", 24, "Male")

dir(person)
```
