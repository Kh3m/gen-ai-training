## Regular Expression

### Methods
- findall
- search
- sub
- split

```python
import re
```

```python
text = """love to code.
I pray God bless my career in AI and Cloud Security"""

print(text)
```

```python
matcher = re.search(r"code", text)
print(matcher)
print(matcher.group())
```

```python
re.findall(r"code|security", text, re.IGNORECASE|re.MULTILINE)
```

```python
re.sub(r"code", "CODE!", text)
```

```python
re.split(r"\.", text)
```

```python
text="""Please contact our regional offices for support. The main support desk is at support@khemshield.com or reach out 
to marketing@gmail.com for queries. Urgent inquiries should call the main office at (123) 456-7890 or 123.456.7890. 
For direct line access, use +234-812-555-0199 or the emergency line +234-987-654-3210.

Emails for specific departments: 
Sales: sales.department@khemshield.com
HR: hr_division@yahoo.com"""

print(text)
```

```python
# Find all email and phone numbers
emails = re.findall(r"\w+@\w+\.\w+", text)
phones = re.findall(r"\+\d{3}-\d{3}-\d{3}-\d{4}", text)

print("Emails: ", emails)
print("Phones: ", phones)
```
