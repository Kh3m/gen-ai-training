```python
# Use a dictionary to count the frequency of elements in list

numbers = [23,6,2,3,6,35,5,6,32,2,42,2,34,5,5,5,5,5,5]
dic = {}
for num in numbers:
    if(num in dic):
        dic[num] = dic[num] + 1
    else:
       dic[num] = 1
        
print(dic)

```
