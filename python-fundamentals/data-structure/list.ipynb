```python
# Basic Syntax: [expression for item in iterable]
lc = [x*2 for x in range(5)]
lc
```

```python
# Filtering Syntax: [expression for item in iterable if condition == True]
lcf = [x*2 for x in range(10) if x % 2 == 0]
lcf
```

```python
# Condition syntax [value if condition == True else alt_value for item in iterable]
lcc = [x if x % 2 == 0 else "-" for x in range(10)]
lcc
```

```python
# Nested comprehension syntax [x for item in iterable]
nlc = [(x, y, z) for x in range(5) for y in range(3) for z in range(2)]
nlc
```

```python
nlst = [3,5,4,6,2,4,23,]

print(nlst.count(4)) # count the occurence of the item
print(nlst.pop(3)) # removes the item at index 3
```
