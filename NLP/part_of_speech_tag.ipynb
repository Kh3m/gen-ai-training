````python
import nltk

nltk.download('averaged_perceptron_tagger_eng')

nltk.pos_tag(["I", "Love", "To", "Code"])
````

````python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

corpus = """
The mysterious old man sat quietly by the shimmering lake, watching as the sun began 
its slow descent behind the distant, jagged mountains. He had spent many years in this 
very spot, thinking about the stories he had heard when he was just a young boy. 
A gentle breeze whispered through the tall pine trees, carrying with it the scent of 
fresh rain and wild lavender. In that moment, everything felt perfectly still, as if 
the world itself was holding its breath.
"""

# Convert to words
words = word_tokenize(corpus.lower())

# Initialize stopwords
stopwords = stopwords.words("english")

up_words = [w for w in words if w not in stopwords]

nltk.pos_tag(up_words)
````
