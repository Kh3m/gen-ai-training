### Following methods and class are used from nltk.tokenize
- sent_tokenize
- word_tokenize
- wordpunt_tokenize
- TreebankWordTokenizer

```python
import nltk

# Make sure the necessary data is downloaded
nltk.download('punkt_tab', quiet=True)
nltk.download('punkt', quiet=True)
```

```python

corpus = """
My name is Abdulkareem and I love AI Security. The reason I need to understand enough workings of ML and AI. 
I pray for Allah's guidance.
"""

print(corpus)
```

```python
documents = nltk.sent_tokenize(corpus)
documents
```

```python
words = [w for x in documents for w in nltk.word_tokenize(x)]
words
```

```python
words = nltk.word_tokenize(corpus)
words
```

```python
from nltk.tokenize import wordpunct_tokenize
print(corpus)
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

print(corpus)
```

```python
treebank_word_tokenizer = TreebankWordTokenizer()

treebank_word_tokenizer.tokenize(corpus)
```
