### Stemming Text Processing
Stemming is the process if reducing a word to its word stem

#### Here we used the following classes:
- PorterStemmer
- RegexpStemmer
- SnowballStemmer

```python
# Could be used in classification problems
# Eating, Eaten -> Eat
# Going, Gone, Goes -> Go

words = ["Eating", "Eats", "Eaten", "Going", "Go", "Gone", "Goes", "Writing", "Writes", "Programming", "Programmer", "Program",
"History", "Historical", "Final", "Finally", "Finalize","Loving", "Love", "Lovely"]
```

```python
words
```

```python
# PorterStemmer
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()

#  porter_stemmer.stem(word)
[ {w: porter_stemmer.stem(w)} for w in words ]
```

### RegexpStemmer
Takes a sinlge regular expression and removes any prefix or siffix that matches the expression.

```python
from nltk.stem import RegexpStemmer

regexp_stemmer = RegexpStemmer('^ing|ing$|s$|e$|able$', min=4)
# regexp_stemmer = RegexpStemmer('ing|s$|e$|able$', min=4) => This removes ing anywhere too


regexp_stemmer.stem("ingEating")
```

## SnowballStemmer 

Still reduces words to its stem and works better than the PortStemmer

```python
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer(language="english")

snowball_stemmer.stem("Congratulations")
```

```python
[{w: snowball_stemmer.stem(w)} for w in words ]
```

```python
print(porter_stemmer.stem("Congratulations"))
print(snowball_stemmer.stem("Congratulations"))
```

```python
print(porter_stemmer.stem("Historical"))
print(snowball_stemmer.stem("Historical"))
```

```python
print(porter_stemmer.stem("fairly"))
print(snowball_stemmer.stem("fairly"))
```
