### Stemming Text Processing
Stemming is the process if reducing a word to its word stem

```python
# Could be used in classification problems
# Eating, Eaten -> Eat
# Going, Gone, Goes -> Go

words = ["Eating", "Eats", "Eaten", "Going", "Go", "Gone", "Goes", "Writing", "Writes", "Programming", "Programmer", "Program",
"History", "Historical", "Final", "Finally", "Finalize","Loving", "Love", "Lovely"]
```

```python
words
```

```python
# PorterStemmer
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

#  stemmer.stem(word)
[ {w: stemmer.stem(w)} for w in words ]
```
